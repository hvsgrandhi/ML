````python
import pandas as pd
````

````python
import nltk
````

````python
nltk.download("stopwords")
nltk.download("punkt")
````

````python
text = "He wanted to escape and he tried his best to do so."
````

````python
from nltk.tokenize import word_tokenize
````

````python
word_tokenize(text)
````

````python
from nltk.corpus import stopwords
````

````python
s_words = stopwords.words('english')
````

````python
cleantext = [word for word in word_tokenize(text) if word not in s_words] #removing the stopwords
````

````python
cleantext
````

````python
from nltk.stem import PorterStemmer
````

````python
ps = PorterStemmer()
````

````python
cleantext = [ps.stem(word) for word in cleantext] #converting  driving, driven, drove to drive
````

````python
cleantext
````

````python
def func(text):
    cleantext = [word for word in word_tokenize(text) if word not in s_words]
    cleantext = [ps.stem(word) for word in cleantext]
    return cleantext
````

````python
func("He wanted to escape and he tried his best to do so")
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
tfidf = TfidfVectorizer(analyzer = func)
````

````python
x = data['Text']
y = data['Label']
````

````python
x_new = tfidf.fit_transform(x)
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, Y_train, Y_test = train_test_split(x_new, y, test_size = 0.25, random_state = 1)
````
